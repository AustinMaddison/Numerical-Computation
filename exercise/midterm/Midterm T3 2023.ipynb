# Problem 1) (10) Taylor

## Find first 3 **non-zero** terms taylor series expansion of
$$ f(x) = \sin\left(\frac{x^2}{2}\right) $$ around $x=0$.

```python
def a_0(f, l):
    xs = np.linspace(0, 2*l, 100)
    ys = [f(x) for x in xs]
    area = np.trapz(ys, xs)
    return (1/(2*l)) * area

def a(f, n, l):
    xs = np.linspace(0, l, 100)
    ys = [f(x) * np.cos(n * np.pi * x / l) for x in xs]
    area = np.trapz(ys, xs)
    return (1/l) * area

def b(f, n, l):
    xs = np.linspace(0, l, 100)
    ys = [f(x) * np.sin(n * np.pi * x / l) for x in xs]
    area = np.trapz(ys, xs)
    return (1/l) * area

def f_series(xs, f, n, l):
    def helper(x, f): 
        result = a_0(f, l)
        for i in range(1, n+1):
            result += a(f, i, l) * np.cos(i * np.pi * x / l) + b(f, i, l) * np.sin(i * np.pi * x / l)
        return result
    ys = [helper(x, f) for x in xs]
    return ys 



def f(x):
    return np.sin(x**2/2)

domain = [0, 1* np.pi]

x = np.linspace(domain[0], domain[1], 1000)

l = 3.8 # eyeballed
print(f'a_0 = {a_0(f, l)}')

n = 3
for i in range(1, n+1):
    print(f'a_{i} = {a(f, i, l)}' '\t' f'b_{i} = {b(f, i, l)}')

y = [ f(xx) for xx in x]
ys = f_series(x, f, 3, l)

plt.plot(x, y, label='f(x)')
plt.plot(x, ys, label='fourier', linestyle='--')

plt.grid()
plt.legend()
plt.show()
```

## Compute the following integral using trapezoid rule

$$
\int_0^1 e^{x} \sin(x^2) dx
$$
use 100 pieces for your integral

```python
def f(x):
    return np.exp(x) * np.sin(x**2)

def trapezoid_integration(f, xs):
    
    def trapezoid_area(y1, y2, dx):
        return y1*dx + ((y2 - y1)*dx)/2
    ys = [f(x) for x in xs]
    
    area = 0
    
    for i in range(len(ys)-1):
        y1, y2, dx = (ys[i], ys[i+1], xs[i+1] - xs[i])
        area += trapezoid_area(y1, y2, dx)
    return area

x = np.linspace(0, 1, 100)
y = [ f(xx) for xx in x]


plt.plot(x, y, label='$e^{x} \sin(x^2)$')
# plt.plot(x, y2, label='$\int_0^1 e^{x} \sin(x^2) dx$')

plt.legend()
plt.grid()

print(f'Area under curve = {trapezoid_integration(f, x)}')
# print(f'Area under curve = {np.trapz(y, x)}') # Check answer
```
