# Numerical Derivative

1) Let $f(x) = x^3$ analytically calculate $f'(0.5)$.

````python
def f(x):
    return x**3
    
def f_prime(x):
    return 3*x**2

f_prime(0.5)
````

2) Use forward finite difference to approximate $f'(0.5)$ with $h=0.1$

````python
def discrete_derivitive(f, x, h):
    return (f(x+h) - f(x))/h

discrete_derivitive(f, 0.5, 0.1)
````

3) Use forward finite difference to approximate $f'(0.5)$ with $h=0.001$

````python
discrete_derivitive(f, 0.5, 0.001)
````

4) Plot $f'(x)$ for $h=0.1$ and $h=0.001$ and analytic one

````python
def plot():
    xs = np.linspace(0, 1, 100)
    gs = [f_prime(x) for x in xs]
    plt.plot(xs, gs, label="analytical", linewidth=3)
    gs = [discrete_derivitive(f, x, 0.1) for x in xs]
    plt.plot(xs, gs, label="h=0.1")
    gs = [discrete_derivitive(f, x, 0.001) for x in xs]
    plt.plot(xs, gs, label="h=0.001", color="red" )
    plt.grid()
    plt.legend()

plot()
````

5) Use centered finite difference to approximate $f'(0.5)$ with $h=0.1$

````python
def discrete_derivitive_center(f, x, h):
    return (f(x+h) - f(x-h))/(2*h)

def plot():
    xs = np.linspace(0, 1, 100)
    gs = [f_prime(x) for x in xs]
    plt.plot(xs, gs, label="analytical", linewidth=3)
    gs = [discrete_derivitive(f, x, 0.1) for x in xs]
    plt.plot(xs, gs, label="h=0.1")
    gs = [discrete_derivitive(f, x, 0.001) for x in xs]
    plt.plot(xs, gs, label="h=0.001", color="red" )
    gs = [discrete_derivitive_center(f, x, 0.1) for x in xs]
    plt.plot(xs, gs, label="h=0.1(centered diff)", color="yellow" )
    plt.grid()
    plt.legend()

plot()
````

6) Write a function that given location $y,x$ and then calculate how "likely" it is that pixel x,y is an edge pixel.

````python
def diff(mario, x,y):
    x = max(
    max(mario[x][y] - mario[x % mario.shape[0]][(y+1) % mario.shape[1]]),
    max(mario[x][y] - mario[x % mario.shape[0]][(y-1) % mario.shape[1]]),
    max(mario[x][y] - mario[(x+1) % mario.shape[0]][y % mario.shape[1]]),
    max(mario[x][y] - mario[(x-1) % mario.shape[0]][y % mario.shape[1]]))
    return x
    
````
